```python
import pandas as pd
```

```python
DATA_PATH = "../data/IOT-temp.csv"
```

```python
df = pd.read_csv(DATA_PATH)
df.head()
```

### 1. Вычислите 5 самых жарких и самых холодных дней за год

```python
out = df[df["out/in"] == "Out"]
tmp = out[["noted_date", "temp"]].groupby(by="noted_date")\
        .mean() \
        .reset_index() \
        .sort_values(by="temp") \
        .reset_index(drop=True)
```

```python
tmp.head(5)
```

```python
tmp.tail(5)
```

### 2 Отфильтруйте out/in = in

```python
df = df[df["out/in"] == "In"]
df.reset_index(inplace=True, drop=True)
df.head()
```

### 3. Поле noted_date переведите в формат ‘yyyy-MM-dd’ с типом данных date;

```python
date = pd.to_datetime(df["noted_date"], format="%d-%m-%Y %H:%M")
df["date"] = date.dt.date

df.head()
```

### 4. Очистите температуру по 5-му и 95-му процентилю

```python
p05 = df.temp.quantile(0.05)
p95 = df.temp.quantile(0.95)

df = df[(df.temp > p05) & (df.temp < p95)]
df.reset_index(inplace=True, drop=True)
df.head()
```
